```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
average_salary = pd.read_csv("data/processed/average_salary_by_year.csv")
salary_experience = pd.read_csv("data/processed/salary_by_year_experience.csv")
salary_companysize = pd.read_csv("data/processed/salary_by_year_company_size.csv")
```

AVERAGE SALARY VISUALIZATION

```python
fig, ax = plt.subplots()

ax.plot(
    average_salary['work_year'],
    average_salary['avg_salary'],
    marker='o',
    label="Average Salary"
)

ax.plot(
    average_salary['work_year'],
    average_salary['median_salary'],
    marker='o',
    linestyle='--',
    label="Median Salary"
)

ax.set_xlabel("Work Year")
ax.set_ylabel("Salary")
ax.set_title("Average vs Median Salary by Year")
ax.legend()

plt.xticks(average_salary['work_year'])
plt.tight_layout()
plt.show()
```

```python
"Over the years, average salaries experienced some fluctuations, likely due to job market conditions or economic disruptions, but overall they have shown substantial growth."
```

AVERAGE SALARY BY EXPERIENCE

```python
# Pivot for plotting
pivot = salary_experience.pivot(index='work_year', columns='experience_level_full', values='avg_salary')

# Plot
fig, ax = plt.subplots(figsize=(10,6))
pivot.plot(ax=ax, marker='o')

ax.set_xlabel("Year")
ax.set_ylabel("Average Salary")
ax.set_title("Average Salary by Experience Level Over Years")
plt.xticks(pivot.index)
plt.legend(title="Experience Level")
plt.tight_layout()
plt.show()
```

```python
"The data reveals a notable increase in average salaries over the years, with the exception of 2021, which may reflect temporary market disruptions. In general, higher experience levels are strongly associated with increased salary levels."
```

SALARY BY COMPANYSIZE

```python
pivot2 = salary_companysize.pivot(index='work_year', columns = 'company_size_full' , values='avg_salary')
fig ,ax = plt.subplots(figsize=(10,6))
pivot2.plot(ax=ax, marker='o')
ax.set_xlabel("Year")
ax.set_ylabel("Average Salary")
ax.set_title(" Average Salary by Company Size Over Years")
plt.xticks(pivot2.index)
plt.legend(title=" Company Size")
plt.show()
```

```python
print('Larger companies generally offer higher salaries, but interestingly, medium-sized companies show the steepest growth in average salary, reflecting a competitive market for experienced talent in growing organizations.')
```
